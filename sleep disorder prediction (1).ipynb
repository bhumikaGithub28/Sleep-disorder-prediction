### About Dataset

The Sleep Health and Lifestyle Dataset comprises 400 rows and 13 columns, covering a wide range of variables related to sleep and daily habits. It includes details such as gender, age, occupation, sleep duration, quality of sleep, physical activity level, stress levels, BMI category, blood pressure, heart rate, daily steps, and the presence or absence of sleep disorders.

### Data Dictionary

### Details about Sleep Disorder Column:

None: The individual does not exhibit any specific sleep disorder.

Insomnia: The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.

Sleep Apnea: The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.

```python
from sklearn import preprocessing 
from sklearn.preprocessing import LabelEncoder
label_encoder = preprocessing.LabelEncoder()
```

```python
vars = ['Gender','Occupation','BMI Category','Sleep Disorder']
for i in vars:
    label_encoder.fit(df[i].unique())
    df[i] = label_encoder.transform(df[i])
    print(i,':',df[i].unique())
```

### Correlation Matrix Heatmap

```python
plt.figure(figsize=(20,16))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Sleep Disorder',axis=1), df['Sleep Disorder'], test_size=0.3)
```

```python
y_train
```

### Model Building

##### For predicting the sleep disorder I will use Decision Tree Classifier with Post Pruning Method 

```python
from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier(max_depth=4)
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(20,16))
tree.plot_tree(treemodel,filled=True)
```

### Decision Tree Model Evaluation

```python
#PREDICTION
y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
#Training Accuracy
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

##### Using Confusion matrix heatmap to visualize the model accuracy

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,cmap='Blues',fmt='g')
```

The diagonal boxes show the count of the true positive results,i.e correct prediction made by the model. The off-diagonal boxes show the count of false positive results,i.e incorrect prediction made by the model.

```python
ax = sns.distplot(y_test,hist = False, color = "r", label="Actual Value")
sns.distplot(y_pred,hist = False,color = "b",label = "Fitted Values",ax=ax)
plt.title('Actual vs Fitted Values for Sleep Disorder Prediction')
plt.xlabel('Sleep Disorder')
plt.ylabel('Proportion of People')
```

```python
print((classification_report(y_pred,y_test)))
```

The model gives pretty decent result with an accuracy of 89% and an average F1 score of 0.88.The model is able to predict the 
sleep disorder with a good accuracy.
